```python
import numpy as np
import matplotlib.pyplot as plt
from percentile import percentile
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10,10))

a = [0,1,2,3]
critical_percentiles = [0, 25, 50, 75, 100]
critical_percentiles_m3 = [0, 38, 62, 88, 100]
for axi, method in zip(axes.reshape(-1), np.arange(1,10)):
    
    # Draw lines
    cp = critical_percentiles
    if method == 3:
        cp = critical_percentiles_m3
    
    for p1, p2 in zip(cp[:-1], cp[1:]):
        percentiles = np.arange(p1+1, p2)
        axi.plot(percentiles, percentile(a, percentiles, method=method), c='k', zorder=2)
    
    # Draw dots
    cp_percentiles = percentile(a, critical_percentiles, method=method)    
    axi.scatter(critical_percentiles, cp_percentiles, c='w', s=70, zorder=3)
    axi.scatter(critical_percentiles, cp_percentiles, c='k', s=20, zorder=4)
    
    axi.grid()
    axi.set_title(f'Method = {method}')
    axi.set_xlabel('Percentile')
    axi.set_ylabel('Returned value')
    
fig.suptitle('Example of Percentile Methods 1-9 for the Array [0,1,2,3]', y = 1.02, fontsize=14)
plt.tight_layout()
```
